```python
courses = []
choice = 'NONE'

def hashline():
  print("-----------------------")

def main_menu():
  print()
  hashline()
  print("     Main Menu")
  hashline()
  print("1. Add Course")
  print("2. Remove Course")
  print("3. View Course List")
  print("4. Exit")
  hashline()
  choice = input("Enter your Choice: ")
  return choice

def view_list():
  print()
  i = 0
  hashline()
  print("     Course List")
  hashline()
  for x in courses:
    print(i, x)
    i = i + 1
  if i == 0:
    print("\n     Empty List\n")
  hashline()

while(choice != 4):
  choice = main_menu()
  if choice == '1':
    course_name = input("Enter your Course Name: ")
    course_code = input("Enter your Course Code: ")
    courses.append(course_code + " - " + course_name)
    print("Course Added Successfully!")
  elif choice == '2':
    view_list()
    id = int(input("Enter ID of any item to remove: "))
    del courses[id]
    print("Item Deleted Successfully!")
  elif choice == '3':
    view_list()
  elif choice == '4':
    break
  else:
    print("Invalid Input! Please select choices between 1-4")
```
